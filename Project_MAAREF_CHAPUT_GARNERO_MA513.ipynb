There are two types of Perceptrons: Single layer and Multilayer.

Single layer Perceptrons can learn only linearly separable patterns.

Multilayer Perceptrons or feedforward neural networks with two or more layers have the greater processing power.

The Perceptron algorithm learns the weights for the input signals in order to draw a linear decision boundary.

This enables you to distinguish between the two linearly separable classes +1 and -1.

## Perceptron Learning rule 

Perceptron Learning Rule states that the algorithm would automatically learn the optimal weight coefficients. The input features are then multiplied with these weights to determine if a neuron fires or not.

```python
img4 = mpimg.imread('perceptron_rule.png')
imgplot2 = plt.imshow(img4)
plt.show()
```

- We observe multiple input signals, if the sum of the input signals exceeds a certain threshold, it either outputs a signal or does not return an output. In the context of supervised learning and classification, this can then be used to predict the class of a sample.

```python
'''Coding section'''
#imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import Perceptron
sns.set()
import matplotlib.patches as mpatches

import warnings
warnings.filterwarnings('ignore')
%matplotlib inline

from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import validation_curve
from sklearn import metrics

import matplotlib.image as mpimg
from sklearn import tree
from sklearn.tree import plot_tree
from io import StringIO

from IPython.display import Image  
from sklearn.tree import export_graphviz
```

```python
#extracting datafram with pandas
df = pd.read_csv('sms_spam_perceptron.csv', delimiter=';')
```

```python
df.describe() #general description of the data
```

```python
df #data
```

```python
samples = df[['sex', 'buy']].values #X -> sample
targets = df["type"].values #Y -> Target   ###defining samples and targets for the perceptron 
```

```python
print('shape of samples', samples.shape)
print('shape of targets',  targets.shape)
```

```python
#test split 
training_samples,   testing_samples, training_targets, testing_targets = train_test_split(samples, targets,random_state=0, test_size=0.3)
```

```python
#shape of split
print(training_samples.shape, testing_samples.shape)
print(testing_targets.shape, testing_targets.shape)
```

```python
#Perceptron Model
Perceptron_classifier = Perceptron()#creating perceptron classifier
```

```python
#fit the data for the decision Tree
Perceptron_classifier.fit(training_samples, training_targets)
```

```python
#testing the prediction Tree
prediction = Perceptron_classifier.predict(testing_samples)
print (prediction [0:20], '(predictions)')
print('testing_targets')
print ( testing_targets[0:20])
print("Perceptron's Accuracy: ", metrics.accuracy_score(testing_targets, prediction))
```

## Pros and Cons of Perceptron 

- Cons : 
Perceptron use a binary classification method as we saw, for complex forms of classification, this can be an obstacle. 
The algorithm auto-correct the mistakes between the result of the activation function and the output,this means that we consider the algorithm robust enough to detect the significant mistakes. This depends also on the difficulty of the data and problem faced.


- Pros : 
Easy to implement, and the results are obtained fast. For our example, we have an 80% accuracy which is correctly acceptable. 
For simple identification of 'ham' or 'spam', we can consider Perceptron effective for this task, because it's an arbitrary choice. The algoritm is trained to detect anomalies, like for the sms_spam file.
